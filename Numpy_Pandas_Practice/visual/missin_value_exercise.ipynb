```python
import pandas as pd
import numpy as np
```

```python
raw_data = {'first_name': ['Jason', np.nan, 'Tina', 'Jake', 'Amy'], 
            'last_name': ['Miller', np.nan, 'Ali', 'Milner', 'Cooze'], 
            'age': [42, np.nan, 36, 24, 73], 
            'sex': ['m', np.nan, 'f', 'm', 'f'], 
            'preTestScore': [4, np.nan, np.nan, 2, 3], 
            'postTestScore': [25, np.nan, np.nan, 62, 70]}

df = pd.DataFrame(raw_data, columns = ['first_name', 'last_name', \
                                       'age', 'sex', 'preTestScore', 'postTestScore']) 
df
```

```python
df.isnull()
#각 항목이 
```

```python
 df.isnull().sum() / len(df)
```

```python
df_no_missing = df.dropna() 
df_no_missing
```

```python
df
```

```python
df_cleaned = df.dropna(how='all') 
df_cleaned
```

```python
df['location'] = np.nan 
df
```

```python
df.dropna(axis=1, how='all')
# how='all'은 모두 nan인 것을 포함한다는 것을 의미한다.
```

```python
df
```

```python
df.dropna(axis=0, thresh=1)
#dropna()는 널이 있는 것을 drop하는 함수
#thresh=1은 np.nan인 것의 갯수가 1이상인 것만 남긴한다.
```

```python
 df.dropna(thresh=5)
#thresh=5은 np.nan인 것의 갯수가 1이상인 것만 남긴다.
```

```python
 df.fillna(0)
#df.fillna(0)은 nan이 있는 곳에 0으로 채운다.
```

```python
df["postTestScore"].fillna( 
    df.groupby("sex")["postTestScore"].transform("mean"), inplace=True)

#fillna은 nan인 값의 항목을 채우는 항목이다. 그런데 df.groupby("sex")["postTestScore"] 여기에\
#값을 채워야 하는데 groupby에 [sex] colum에 값이 nan이라 값을 채울 수 없다.
df
```
