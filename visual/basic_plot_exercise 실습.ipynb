```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams["figure.figsize"] = (10,5)
#뒤에 캔버스(figure은 캔버스 역할을 한다)의 사이즈, 비율을  (10,5)로 만든다
#위에 있는 예시는 figure을 생성해야하지만, pyplot 객체를 사용시, 기본 figure에 그래프가 그려진다.

x = range(100)
y=[value**2 for value in x]

plt.plot(x,y)
plt.savefig('test.png')
```

```python
import numpy as np

X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1]
#value를 입력하면 값을 입력하면 실수를 greed라고 생각한다.

X_2 = range(100)
Y_2 = [np.sin(value) for value in X_2]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)

plt.show()
```

```python
fig = plt.figure()# figure 반환
fig.set_size_inches (10,5) #figure 크기지정

ax_1 = fig.add_subplot(1,2,1)#figure을 행1개, 열2개로 나누고, 1번째에 넣는다. 
ax_2 = fig.add_subplot(1,2,2)#figure을 행1개, 열2개로 나누고, 2번째에 넣는다.

ax_1.plot(X_1, Y_2, c="b") #색깔은 blue
ax_2.plot(X_2, Y_2, c="g") #색깔은 green
plt.show()
```

# Set Color

```python
X_1 = range(100)
Y_1 = [value for value in X_1]

X_2 = range(100)
Y_2 = [value + 50 for value in X_2]

plt.plot(X_1, Y_1, color = "#000000")
plt.plot(X_1, Y_1, c = "c") #color은 c로 쓸 수 있다.

plt.show()
```

# Set linestyle

```python
plt.plot(X_1, Y_1, c="b", linestyle="dashed") 
plt.plot(X_2, Y_2, c="r", ls="dotted")#linestyle은 ls로 쓸 수 있다.

plt.show()
```

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed") 
plt.plot(X_2, Y_2, color="r", linestyle="dotted")
plt.title("Two lines")
plt.show()
```

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed") 
plt.plot(X_2, Y_2, color="r", linestyle="dotted")

plt.title('$y = \\frac{ax + b}{test}$')  #타이틀의 이름을 latex타입(수식)으로 표현 가능하다
plt.show()
```

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed") 
plt.plot(X_2, Y_2, color="r", linestyle="dotted")

plt.title('$y = ax+b$')

plt.xlabel('$x_line$') #x축에 이름을 붙일 수 있다(latex 타입으로).
plt.ylabel('y_line') #y축에 이름을 붙일 수 있다.

plt.show()
```

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed") 
plt.plot(X_2, Y_2, color="r", linestyle="dotted")
plt.text(50, 70, "Line_1") 
plt.annotate( 'line_2', xy=(50, 100), xytext=(30, 110), arrowprops=dict(facecolor='black', shrink=0.1))
plt.title('$y = ax+b$') 
plt.xlabel('$x_line$') 
plt.ylabel('y_line')
plt.show()
```

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed", label='line_1') 
plt.plot(X_2, Y_2, color="r", linestyle="dotted", label='line_2') 
plt.legend(shadow=True, fancybox=False, loc="upper right")
#범례를 표시할 수 있다, loc로 위치 지적할 수 있다.
#shadow는 그림자, fancybox는 그냥 멋있게 만드는 것
plt.title('$y = ax+b$') 
#그래프 이름
plt.xlabel('$x_line$') 
#x축 이름
plt.ylabel('y_line')
#y축 이름
plt.show()
```

```python
N = 50 
x = np.random.rand(N) 
y = np.random.rand(N) 
colors = np.random.rand(N) 
area = np.pi * (15 * np.random.rand(N))**2 
plt.scatter(x, y, s=area, c=colors, alpha=0.5) 
plt.show()
```

# Bar chart 

```python
 data = [[5., 25., 50., 20.], 
         [4., 23., 51., 17], 
         [6., 22., 52., 19]]
    
X = np.arange(0,8,2)
```

```python
plt.bar(X + 0.00, data[0], color = 'b', width = 0.50) 
plt.bar(X + 0.50, data[1], color = 'g', width = 0.50) 
plt.bar(X + 1.0, data[2], color = 'r', width = 0.50) 
plt.xticks(X+0.50, ("A","B","C", "D")) 

plt.show()
```

```python
 data = np.array([[5., 25., 50., 20.], 
                  [4., 23., 51., 17], 
                  [6., 22., 52., 19]])
color_list = ['b', 'g', 'r'] 
data_label = ["A","B","C"] 
X = np.arange(data.shape[1]) 
X
```

```python
 data = np.array([[5., 5., 5., 5.], 
                  [4., 23., 51., 17], 
                  [6., 22., 52., 19]])
    
for i in range(3): 
    plt.bar(X, data[i], bottom = np.sum(data[:i], axis=0), 
            color = color_list[i], label=data_label[i]) 
plt.legend() 
plt.show()
```

```python
A = [5., 30., 45., 22.] 
B = [5, 25, 50, 20]
X = range(4)

plt.bar(X, A, color = 'b') 
plt.bar(X, B, color = 'r', bottom = 60) 
plt.show()
```

```python
women_pop = np.array([5, 30, 45, 22]) 
men_pop = np.array([5, 25, 50, 20]) 
X = np.arange(4)

plt.barh(X, women_pop, color = 'r') 
plt.barh(X, -men_pop, color = 'b') 
plt.show()
```
